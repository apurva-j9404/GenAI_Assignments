```python
KAIZEN_PROMPT = """
You are a Japanese-method based study assistant.

Follow this learning cycle strictly:

1. Simple Explanation (2-3 lines)
2. Feynman Explanation (explain like to a beginner)
3. Key Points (bullet points)
4. Active Recall Question
5. Kaizen Tip (small improvement suggestion)

User Question:
{question}

Format clearly with headings.
"""
```

```python
from google.colab import userdata
import google.generativeai as genai

api_key = userdata.get("geminiAPI")
genai.configure(api_key=api_key)

model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
# Build Response Function

def kaizen_answer(question):
    prompt = KAIZEN_PROMPT.format(question=question)
    response = model.generate_content(prompt)
    return response.text
```

```python
# Chat Loop

print("🇯🇵 Kaizen Learning Cycle Chatbot")
print("Type 'exit' to quit")

while True:
    user_input = input("\nYou: ")
    if user_input.lower() == "exit":
        print("Keep improving. Sayonara 👋")
        break

    output = kaizen_answer(user_input)
    print("\nKaizenBot:\n")
    print(output)
```
