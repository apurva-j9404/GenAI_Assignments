```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
#Step 1: Create Sample Data

np.random.seed(0)

data1 = np.random.normal(loc=0, scale=1, size=(100,2))
data2 = np.random.normal(loc=5,scale=1,size=(100,2))
data3 = np.random.normal(loc=10,scale=1,size=(100,2))

X = np.vstack((data1,data2,data3))
```

```python
# Step2: Build GMM model

gmm = GaussianMixture(n_components=3)
gmm.fit(X)
```

```python
# Step3 : Predict cluster labels
labels = gmm.predict(X)
```

```python
#Step4 : Generate new samples
new_samples, _ = gmm.sample(50)
```

```python
#Step5: Visualization
plt.scatter(X[:,0],X[:,1],c=labels)
plt.scatter(new_samples[:,0], new_samples[:,1], marker='x')
plt.title("Gaussian Mixture Model")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
