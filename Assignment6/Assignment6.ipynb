```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
```

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
def generate_image(prompt):

    image = pipe(prompt).images[0]

    plt.figure(figsize=(6,6))
    plt.imshow(image)
    plt.axis("off")

    return image
```

```python
generate_image("A lotus blooming in waterbody")

generate_image("A women painting a masterpiece in a old art studio")

generate_image("A magical forest with glowing mushrooms and fireflies")
```

```python
image = generate_image("Eileen Gu skiing on snow covered landscape")

image.save("generated_image.png")
```
